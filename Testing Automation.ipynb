```python

def automateSmallInventory(inventory, realTimeInventory):
    orig_df = pd.read_csv(inventory)
    orig_df.drop(columns = ['Unnamed: 7'])
    realTime_df = pd.read_csv(realTimeInventory)
    realTime_df = realTime_df[['Name', 'Location Available', 'Location']]
    realTime_df = realTime_df[realTime_df['Location'] == 'KFHQ']
    realTime_df['Name'] = realTime_df['Name'].astype(str)
    orig_df['SKU'] = orig_df['SKU'].astype(str)
    realTime_df.rename(columns = {'Name':'SKU'}, inplace = True)
    new_df = pd.merge(orig_df, realTime_df, on = 'SKU', how = 'left')
    chino = new_df['Chino']
    supply = new_df['Location Available']

    #somehow we going into all the rows

    for x in range(len(chino)):
        if math.isnan(supply[x]) == False:

            if supply[x] <= 40:
                chino[x] = 0
            elif supply[x] <= 60:
                chino[x] = 10
            elif supply[x] <= 100:
                chino[x] = 20
            else:
                chino[x] = int(supply[x] * .3)
                
    new_df['Chino'] = chino
    new_df = new_df.iloc[:,0:7]
    new_df.to_excel('.\\Homepop\\revisedHomePOP_' + str(date.today()) + '.xlsx')
```

```python
def automateKirkland(kirk_inv, curr_inventory):
    kirkland_df = pd.read_excel(kirk_inv)
    kirkland_df.rename(columns = {'Vendor Part Number ':'SKU'}, inplace = True)
    kirkland_df['SKU'] = kirkland_df['SKU'].astype(str)
    
    realTime_df = pd.read_csv(curr_inventory)
    realTime_df = realTime_df[['Name', 'Location Available', 'Location']]
    realTime_df = realTime_df[realTime_df['Location'] == 'KFHQ']
    realTime_df['Name'] = realTime_df['Name'].astype(str)

    realTime_df.rename(columns = {'Name':'SKU'}, inplace = True)
    updated_kirkland = pd.merge(kirkland_df, realTime_df, on = 'SKU', how = 'left')
    
    supply = updated_kirkland['Location Available']
    chino = updated_kirkland['Inventory']
    for x in range(len(chino)):
        if math.isnan(supply[x]) == False:

            if supply[x] <= 40:
                chino[x] = 0
            elif supply[x] <= 60:
                chino[x] = 10
            elif supply[x] <= 100:
                chino[x] = 20
            else:

                chino[x] = int(supply[x] * .3)

    updated_kirkland['Inventory'] = chino
    updated_kirkland = updated_kirkland.iloc[: , 0:3]
    updated_kirkland.to_excel('.\\Kirkland\\updated_kirkland_' + str(date.today()) + '.xlsx')
```

```python
kirk_inv = "Kirkland's Inventory.xlsx"
curr_inventory = 'KinfineRealTimeInventoryResults 10-25-2021.csv'
automateKirkland(kirk_inv,curr_inventory)    
```

```python
def automateNFM(inventory, realTimeInventory):
    
    realTime_df = pd.read_csv(realTimeInventory)
    realTime_df = realTime_df[['Name', 'Location Available', 'Location']]
    realTime_df = realTime_df[realTime_df['Location'] == 'KFHQ']
    realTime_df['Name'] = realTime_df['Name'].astype(str)

    realTime_df.rename(columns = {'Name':'SKU'}, inplace = True)
    
    
    nfm_df = pd.read_excel(inventory)
    nfm_df.rename(columns = {'20210329':'SKU'}, inplace = True)
    updated_nfm = pd.merge(nfm_df, realTime_df, on = 'SKU', how = 'left')
    stock = updated_nfm['CA']
    availability = updated_nfm['Chino']
    supply = updated_nfm['Location Available']

    for x in range(len(stock)):
        if math.isnan(supply[x]) == False:
            #change supply
            if supply[x] <= 40:
                stock[x] = 0
            elif supply[x] <= 60:
                stock[x] = 10
            elif supply[x] <= 100:
                stock[x] = 20
            else:
                stock[x] = int(supply[x] * .3)

            #change availability
        if availability[x] != 'DI':
            if stock[x] == 0:
                availability[x] = 'OS'
            else:
                availability[x] = 'AV'
    updated_nfm['CA'] = stock
    updated_nfm['Chino'] = availability
    updated_nfm = updated_nfm.iloc[:,0:11]
    updated_nfm.to_excel('.\\NFM\\updated_NFM_' + str(date.today()) +'.xlsx', index = False)
```

```python
realTime = 'KinfineRealTimeInventoryResults 10-25-2021.csv'
inventory = 'nfmInventory.xlsx'
automateNFM(inventory, realTime)
```

```python
def automateHomeDepot(inventory, realTime):
    #load real time data
    realTime_df = pd.read_csv(realTime)
    realTime_df = realTime_df[['Name', 'Location Available', 'Location']]
    realTime_df = realTime_df[realTime_df['Location'] == 'KFHQ']
    realTime_df['Name'] = realTime_df['Name'].astype(str)

    realTime_df.rename(columns = {'Name':'SKU'}, inplace = True)

    homedepot_df = pd.read_excel(inventory)

    homedepot_df.rename(columns = {'VENDOR SKU':'SKU'}, inplace = True)
    updated_homedepot = pd.merge(homedepot_df, realTime_df, on = 'SKU', how = 'left')
    stock = updated_homedepot['QTY']
    availability = updated_homedepot['AVAILABLE']
    supply = updated_homedepot['Location Available']

    

    for x in range(len(stock)):
        if math.isnan(supply[x]) == False:
            #change supply
            if supply[x] <= 40:
                stock[x] = 0
            elif supply[x] <= 60:
                stock[x] = 10
            elif supply[x] <= 100:
                stock[x] = 20
            else:
                stock[x] = int(supply[x] * .3)

            #change availability
        if stock[x] == 0:
            availability[x] = 'No'
        else:
            availability[x] = 'Yes'



    updated_homedepot['AVAILABLE'] = availability
    updated_homedepot['QTY'] = stock
    updated_homedepot = updated_homedepot.iloc[:,0:45]

    updated_homedepot.to_csv('.\\HomeDepot\\updatedHomeDepot_' + str(date.today()) + '.csv')
```

```python
inventory = 'HomeDepotInventory1.xlsx'
realTime = 'KinfineRealTimeInventoryResults 10-25-2021.csv'
automateHomeDepot(inventory, realTime)
```
